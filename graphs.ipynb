```python
# Import matplotlib, numpy and math 
import matplotlib.pyplot as plt 
import numpy as np 
import math 
  
x = np.linspace(-4, 4, 50) 
z = 1/(1 + np.exp(-x))
l =[-4,-3,-2,-1,0,1,2, 3, 4]
h = [-1, -1, -1, -1,-1, 1, 1, 1, 1]

# Compute ReLU values for corresponding x
relu = np.maximum(0, x)
tanh = np.tanh(x)

plt.plot(x, relu, label= "ReLU")
plt.plot(x, tanh, label="Tanh")
plt.step(l, h, label="Step") 
plt.xlabel('Indata')
plt.ylabel('Utdata')
plt.legend()
  
plt.show()

```
